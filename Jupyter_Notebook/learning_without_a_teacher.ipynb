```python
from zipfile import ZipFile

with ZipFile( 'data.zip', 'r') as zip:
  zip.extractall()
  print('Done')
```

```python
import pandas as pd
import numpy as np

df_2d = pd.read_csv("data/data_2d.csv")
df_mnist = pd.read_csv("data/mnist.csv")

df_2d_numpy = df_2d.to_numpy()
df_mnist_numpy = df_mnist.to_numpy()


from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

kmeans_2d = KMeans(n_clusters=2)
kmeans_2d.fit(df_2d)
clusters_2d = kmeans_2d.predict(df_2d)
centroids_2d = kmeans_2d.cluster_centers_


kmeans_mnist = KMeans(n_clusters=10)
kmeans_mnist.fit(df_mnist)
clusters_mnist = kmeans_mnist.predict(df_mnist)

print("Cluster assignments for some examples in MNIST dataset:")
for i in range(5):
    print(f"Example {i+1}: Cluster {clusters_mnist[i]}")
```

```python
from sklearn.decomposition import PCA

kmeans_2d = KMeans(n_clusters=3)
clusters_2d = kmeans_2d.fit_predict(df_2d)

plt.scatter(df_2d_numpy[:, 0], df_2d_numpy[:, 1], c=clusters_2d, cmap='viridis')
plt.title('Clustering Result for 2D Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.show()

pca = PCA(n_components=2)
mnist_2d = pca.fit_transform(df_mnist)

kmeans_mnist = KMeans(n_clusters=10)
clusters_mnist = kmeans_mnist.fit_predict(mnist_2d)

plt.scatter(mnist_2d[:, 0], mnist_2d[:, 1], c=clusters_mnist, cmap='viridis')
plt.title('Clustering Result for MNIST Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
